```python
# Importar las librerias
import numpy as np
import pandas as pd
```

```python
# Cargar el Dataset de Raisin Y Wine
# archivo_csv = 'C:/Users/S ALBERT FC/Documents/ESCOM/5째 semestre/Aprendizaje Maquina/wine.csv'
archivo_csv = 'C:/Users/S ALBERT FC/Documents/ESCOM/5째 semestre/Aprendizaje Maquina/raisin.csv'
data = pd.read_csv(archivo_csv, delimiter=',')
```

```python
# Funcion para separar el dataset en entrenamiento y prueba
def split_data(data, iteracion):
    indices = data.index.tolist()  # Lista de indices de las filas en el conjunto de datos
    test_data = data.iloc[[iteracion]]  # Seleccionar una fila (el patr처n para prueba)
    train_data = data.drop(iteracion)  # Todas las filas excepto la de test (para entrenamiento)
    return train_data, test_data
```

```python
# Funcion para hacer el metodo de validacion para el dataset
def leave_one_out(data, iteracion):
    train_data, test_data = split_data(data, iteracion)  # Obtener los conjuntos de entrenamiento y prueba
    print("\nMUESTRA PARA PRUEBA:\n", test_data)
    print("MUESTRAS PARA ENTRENAMIENTO:\n", train_data)
```

```python
# Ejecutar el algoritmo Leave-One-Out
num_patrones = len(data)
for iteracion in range(num_patrones):
    leave_one_out(data, iteracion)
```
